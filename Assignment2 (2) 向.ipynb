## (1) 按照Orderlist里的顺序

```python
def calculate_total_distance(distance_matrix, orders, allocation):
    total_square = 0
    starting_point = "Packaging"
    
    # 遍历DataFrame的每一行
    for index, order in orders.iterrows():  # 使用.iterrows()获取索引和行
        current_position = starting_point
        # visited = set()  # 使用集合跟踪访问过的货架编号
        
        # 遍历订单中的每个产品组
        for product_group in order:  # 'order' 是一个Series

            # 忽略值为0的产品组
            if product_group > 0: #and product_group not in visited:
                # 索引该product所在的货架编号
                product_group_row = allocation[allocation['Product Group'] == product_group]
                shelf_number = product_group_row.index[0]

                # 累加距离
                total_square += distance_matrix[current_position][shelf_number]
                current_position = shelf_number
                # visited.add(product_group)

        total_square += distance_matrix[current_position][starting_point]

    # 计算总距离
    total_distance = total_square * 3

    return total_square, total_distance


print(calculate_total_distance(DistanceMatrix, Orders, Allocation))
```

## (1) Greedy

```python
def calculate_total_distance(distance_matrix, orders, allocation):
    total_square = 0
    starting_point = "Packaging"

    # 遍历DataFrame的每一行
    for index, order in orders.iterrows():
        current_position = starting_point
        # visited = set()

        # 定义剩余产品组集合
        remaining_products = set(order[order > 0])

        while remaining_products:
            # 计算当前位置到各个未访问产品组的距离
            for product in remaining_products:

                product_group_row = allocation[allocation['Product Group'] == product]
                shelf_number = product_group_row.index[0]

                distances_to_remaining_products = {}
                distances_to_remaining_products[product] = distance_matrix[current_position][shelf_number]
            
            # 选择最近的产品组
            nearest_product = min(distances_to_remaining_products, key=distances_to_remaining_products.get)
            nearest_product_row = allocation[allocation['Product Group'] == nearest_product]
            nearest_shelf = nearest_product_row.index[0]

            # 移动到最近的货架
            total_square += distances_to_remaining_products[nearest_product]
            current_position = nearest_shelf

            # 标记当前产品组为已访问
            # visited.add(nearest_product)
            remaining_products.remove(nearest_product)

        # 返回起始点
        total_square += distance_matrix[current_position][starting_point]

    total_distance = total_square * 3 

    return total_square, total_distance

print(calculate_total_distance(DistanceMatrix, Orders, Allocation))
```

## (2) Construction Heuristic

```python
# Create a problem 

constr_heuristic = xp.problem(name='Construction Heuristic')
```

```python
# Define the matrix dimension for the decision variable (product group number x shelf number)
rows = 90
cols = 96

# Create the x variable to store the allocation decisions (使用列表推导式创建一个90x96的二维列表，每个元素都是一个二进制变量)
x = [[xp.var(vartype=xp.binary) for j in range(cols)] for i in range(rows)]

# Add the variables to the problem
for row in x:
    constr_heuristic.addVariable(row)
```
